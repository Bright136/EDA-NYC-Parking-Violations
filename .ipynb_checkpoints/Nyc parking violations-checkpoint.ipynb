<a id='Cleaning Data'></a>
## Cleaning Data

```python
# since the summons number and the violation code are a unique values convert to strings
df['Summons Number'] = df['Summons Number'].astype(str)
df['Violation Code'] = df['Violation Code'].astype(str)
```

```python
# change issue date inot a datetime data type
df['Issue Date'] = pd.to_datetime((df['Issue Date']))
```

```python
df = df.dropna(subset=['Plate ID','Registration State','Street Name'])
```

```python
df = df[df['Registration State'] != '99']
df['Registration State'].unique()
```

```python
df[df['Vehicle Year'] == 0]
```

```python
df = df[(df['Vehicle Year'] != 0) & (df['Vehicle Year'] <= 2021)]
df
```

```python
# check the unique values of vehicles 
print(df['Vehicle Make'].nunique())
df['Vehicle Make'].unique()
```

```python
# display issue date columns 
print(df['Issue Date'].dtype)
df['Issue Date']
```

```python
colors = df['Vehicle Color'].str.upper().replace(np.nan, 'OTHER')
df.loc[:, 'Vehicle Color'] = colors
```

```python
vehicle_colors =  df['Vehicle Color']
vehicle_colors =  df['Vehicle Color'].unique()
vehicle_colors.sort()
vehicle_colors
```

```python
df = df[df['Plate Type'] != '999']
df.reset_index()
```

```python
upper_case = df['Vehicle Make'].str.upper().str.strip().replace(np.nan, 'UNKNOWN')
df.loc[:, 'Vehicle Make'] = upper_case
```

```python
unique_veh_mk = df['Vehicle Make'].unique()
unique_veh_mk.sort()
unique_veh_mk
```

```python
vehicle_with_start_t = df[df['Vehicle Make'].str.match('[Tt]')]
vehicle_with_start_t['Vehicle Make'].unique()
```

```python
toyota = df[df['Vehicle Make'].str.match('[Tt]')]
```

<a id='Analyzing and Visualizing'></a>
## Analyzing and Visualizing

```python
base_color = sb.color_palette()[0]
```

```python
bins = np.arange(0, df['Vehicle Year'].max()+1, 1)

plt.figure(figsize=[8, 5])
plt.hist(data = df, x = 'Vehicle Year', bins=bins)
plt.title('Distribution of Years')
plt.xlabel('years')
plt.ylabel('Years')
plt.axis([1976, 2022, 0, 6500])
plt.show()
```

>- The above distribution is skewed negatively. The distribution is concerntrated between the years 2015 and 2020

```python
plt.figure(figsize=(25, 5))
sb.countplot(data=df, x='Vehicle Year', color=base_color);
plt.title('Vehicle Year');
```

>- 2019 had the  hisghest count. In 2019 more vehicle had more violation tickets done the rest of the year

```python
# check the top five car with the highest violations tickets from 1976 to 2021
top_5_cars = df['Vehicle Make'].value_counts().head()
top_5_cars.plot(kind='bar', title='Top 5 Vehicle Make', xlabel='Vehicle Make', ylabel='Number of Violation Tickets');

```

>- The top 5 cars with the highest violations tickets are Honda, Toyota, Ford, Nissan and Chevrolet

```python
# create a new york dataframe
new_york_df = df[df['Registration State'] == 'NY'] # a dataframe with new york as the registration state
```

```python
new_york_df
```

```python
# check the top 5 cars in new yor with highest total violation tickets
new_york_top_cars = new_york_df['Vehicle Make'].value_counts().head()
new_york_top_cars.plot(kind='bar', title='Top 5 Vehicle Make in New York', x='Vehicle Make', y='Number of Violation Tickets')
```

```python
# create a data frame  where the toyota is the only vehicle make
toyota_df = df[(df['Vehicle Make'] == 'TOYOT')]

# group by vehicle and find count 
toyota_by_year = toyota_df.groupby(['Vehicle Year'])['Vehicle Make'].count()
plt.plot(toyota_by_year)
plt.title('Trend of Toyota Vehicles over the Years')
plt.ylabel('Years')
plt.xlabel('umber of Vehicles')
plt.show()
```

>- The plot above show the trend of toyota cars that receive violations tickets over the years

```python
df.info()
```

```python
# created a new colum for name of day for issue date
df['Issue Day'] = df['Issue Date'].dt.day_name().copy()
df = df.sort_values(by=['Issue Day'],  ascending=True)
```

```python
days = df['Issue Day'].value_counts().sort_values()
plt.plot(days)
plt.title('Violation ticket per day')
plt.xlabel('Day Name')
plt.ylabel('Number of Violation tickets')
plt.show()
```

>- The plot  above clearly shows more violation tickets were issued Thursday than the rest of the days
